```python
import pandas as pd
from pathlib import Path

eeg_path = Path(r'D:\Kartik\IIITB\Study\Sem 3\PE\Kedro\kartik-eeg\data\01_raw\dataset.csv')
stamps_path = Path(r'D:\Kartik\IIITB\Study\Sem 3\PE\Kedro\kartik-eeg\data\02_intermediate\video_data\16_to_30\trainingStamps.csv')

filtered_eeg = pd.DataFrame()
for user in range(1,31):
    eeg = pd.read_csv(eeg_path, parse_dates=['ftime'])
    stamps = pd.read_csv(stamps_path)
    user_stamps=stamps[stamps['participant']==user]
    user_stamps['start']=pd.to_datetime(user_stamps['start'], format='%H:%M:%S')
    user_stamps['end']=pd.to_datetime(user_stamps['end'], format='%H:%M:%S')
    user_stamps.drop(columns=['participant','task','file'], inplace=True)
    for stamp in user_stamps.iterrows():
        temp_df = eeg[(stamp[1].iloc[0]<=eeg['ftime']) & (eeg['ftime']<=stamp[1].iloc[1]) & (eeg['participant']==user)]
        filtered_eeg= pd.concat([filtered_eeg, temp_df])
```

```python
import os
from pathlib import Path
from re import I
import pandas as pd

path=Path(r'D:\Kartik\IIITB\Study\Sem 3\PE\Kedro\kartik-eeg\data\02_intermediate\video_data\1_to_15\taskwise_vid_stamps')
df=pd.DataFrame(columns=['task','file','start','end'])
for file in os.listdir(path):
    if file.endswith('.csv'):
        df_file=pd.read_csv(path/file)
        df_file['participant']=str(int(file[:-4]))
        df=pd.concat([df,df_file],ignore_index=True, sort=True)
        # print(len(df_file), len(df))
save_path=Path(r'D:\Kartik\IIITB\Study\Sem 3\PE\Kedro\kartik-eeg\data\02_intermediate\video_data\1_to_15')
df=df.reindex(columns=['participant','task','file','start','end'])
df.to_csv(save_path/"trainingVidStamps.csv", index=False)
df
```
